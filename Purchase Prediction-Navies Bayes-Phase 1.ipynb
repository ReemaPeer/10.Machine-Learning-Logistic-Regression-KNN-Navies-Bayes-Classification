```python
# Importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the dataset

dataset = pd.read_csv("Social_Network_Ads.csv")
dataset
```

```python
# Converting the Nominal data to Numeric Data

dataset = pd.get_dummies(dataset, drop_first = True)
dataset
```

```python
# Getting the column names of the dataset

dataset.columns
```

```python
# Dropping the User ID column from the dataset

dataset = dataset.drop('User ID', axis = 1)
dataset
```

```python
# Getting the distinct output counts

dataset['Purchased'].value_counts()
```

```python
# Splitting the input and the output data

independent = dataset[['Age', 'EstimatedSalary', 'Gender_Male']]
independent
```

```python
dependent = dataset[['Purchased']]
dependent
```

```python
# Getting the total rows and columns in the input

independent.shape
```

```python
#Splitting the train and thetest set data

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)
```

```python
# Creating the Gaussian Navies Bayes Model

from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train,y_train)

# Predicting the output for the test set

y_pred = classifier.predict(x_test)

# Evaluating the model using confusion matrix

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Printing the Classification Report

from sklearn.metrics import classification_report
c_report = classification_report(y_test, y_pred)
print(c_report)
```

```python
# Creating the Multinomial Navies Bayes Model

from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(x_train,y_train)

# Predicting the output for the test set

y_pred = classifier.predict(x_test)

# Evaluating the model using confusion matrix

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Printing the Classification Report

from sklearn.metrics import classification_report
c_report = classification_report(y_test, y_pred)
print(c_report)
```

```python
# Creating the Bernoulli Navies Bayes Model

from sklearn.naive_bayes import BernoulliNB
classifier = BernoulliNB()
classifier.fit(x_train,y_train)

# Predicting the output for the test set

y_pred = classifier.predict(x_test)

# Evaluating the model using confusion matrix

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Printing the Classification Report

from sklearn.metrics import classification_report
c_report = classification_report(y_test, y_pred)
print(c_report)
```

```python
# Creating the Categorical Navies Bayes Model

from sklearn.naive_bayes import CategoricalNB
classifier = CategoricalNB()
classifier.fit(x_train,y_train)

# Predicting the output for the test set

y_pred = classifier.predict(x_test)

# Evaluating the model using confusion matrix

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

# Printing the Classification Report

from sklearn.metrics import classification_report
c_report = classification_report(y_test, y_pred)
print(c_report)
```
